```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import ast
```

```python
csv_filepaths = [
    "data\\audio\\csv\\out_0_last.csv",
    "data\\audio\\csv\\out_1_last.csv",
    "data\\audio\\csv\\out_2_last.csv",
    "data\\audio\\csv\\out_3_last.csv"
]
```

```python
dfs = [pd.read_csv(csv_file) for csv_file in csv_filepaths]

print(dfs[0].head())
```

```python
# plot
fig, ax = plt.subplots()
for channel, df in enumerate(dfs):
    ax.plot(df['aTimeStamp'], df['loudness'], label=f'channel {channel}')

# plot mean loudness
plt.title('Loudness')
plt.xlabel('Time (s)')
plt.ylabel('Loudness')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# audio data
fig, ax = plt.subplots()

for channel, df in enumerate(dfs):
    ax.plot(df['aTimeStamp'], df['audio'], label=f'channel {channel}')
plt.title('Audio Data')
```
